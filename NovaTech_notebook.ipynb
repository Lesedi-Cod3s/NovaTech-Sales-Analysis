```python
# Checking duplicates deeper to understand severity
dup_prod_id = df.groupby('Product ID')['Product ID'].value_counts().nlargest(10)
dup_prod_id
```

```python
# A function to create unique IDs
def make_ids_unique(df, identifier):
    """
    Make duplicate IDs unique by appending a count.
    
    Args:
        df (pd.DataFrame): Input DataFrame.
        identifier (str): Column name containing IDs.
    
    Returns:
        pandas.Series: Series with unique IDs.
    """
    # Create a copy of the original DataFrame
    df_unique = df.copy()
    
    # Create a series with duplicates flagged
    duplicates = df_unique[identifier].duplicated(keep=False)
    
    # Create a group counter for duplicates
    if duplicates.any():
        # Add a suffix to duplicate IDs
        df_unique.loc[duplicates, identifier] = (
            df_unique.loc[duplicates, identifier] + '_' + 
            df_unique.loc[duplicates].groupby(identifier).cumcount().astype(str)
        )
    
    return df_unique[identifier]

# Correct way to use the function
df['Customer ID'] = make_ids_unique(df, 'Customer ID')
df['Order ID'] = make_ids_unique(df, 'Order ID')
df['Product ID'] = make_ids_unique(df, 'Product ID')

# Checking for Uniqueness amongst IDs
print(f"Customer ID : {df['Customer ID'].duplicated().sum()}")
print(f"Order ID : {df['Order ID'].duplicated().sum()}")
print(f"Product ID : {df['Product ID'].duplicated().sum()}")
print(f"Row ID : {df['Row ID'].duplicated().sum()}")
```
